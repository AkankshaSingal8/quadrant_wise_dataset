```python
import os

def rename_folder(folder_path):
    try:
        for i in range (len(os.listdir(folder_path))):
            # Get the parent directory of the folder
            parent_dir = os.path.join(folder_path, f"{i + 1}")
            
            # Construct the new folder path
            new_folder_path = os.path.join(folder_path, str(i + 1 + 55))
            
            # Rename the folder
            os.rename(parent_dir, new_folder_path)
            print(f"Folder renamed: {parent_dir} -> {new_folder_path}")
            
    except Exception as e:
        print(f"An error occurred: {e}")

# Usage Example
folder_path = "goal_height_5_init3"
rename_folder(folder_path)
```

```python
import os

def rename_files_in_folder(folder_path):
    
    files = os.listdir(folder_path)
    for bag_folder in range (len(files)):
        imgs = [f for f in os.listdir(f"{folder_path}/{bag_folder + 1}") if f.endswith('.png')]
        # print(bag_folder + 1, imgs)
        for img_no in range (len(imgs)):
            old_file_path = f"{folder_path}/{bag_folder + 1}/Image_bag{bag_folder + 1}_img{img_no + 1}.png"
            new_file_path = f"{folder_path}/{bag_folder + 1}/Image{img_no + 1}.png"
            os.rename(old_file_path, new_file_path)
            print(f"Renamed: {old_file_path} -> {new_file_path}")

# Usage Example
folder_path = "goal_height_5_init3"

rename_files_in_folder(folder_path)
```
